```python
def es_triangulo_rectangulo(a, b, c):
  """
  Dados tres números no negativos representando los lados de un triángulo,
  retornamos True si este es rectángulo, False en otro caso.
  """
  return (
      a * a + b * b == c * c or
      a * a + c * c == b * b or
      b * b + c * c == a * a
  )
```

```python
print(es_triangulo_rectangulo(5, 4, 3))
print(es_triangulo_rectangulo(5, 12, 13))
print(es_triangulo_rectangulo(1, 2, 3))
```

```python
import math

def kerry(a, b, c):
  """
  Dados tres números no negativos representando los lados de un triángulo,
  retornamos True si este es rectángulo, False en otro caso.
  """
  return (
      math.sqrt(a * a + b * b) == c or
      math.sqrt(a * a + c * c) == b or
      math.sqrt(b * b + c * c) == a
  )
```

```python
print(kerry(500000000000000000000000000000000, 400000000000000000000000000000000, 300000000000000000000000000000000))
print(kerry(5, 12, 13))
print(kerry(1, 2, 3))
```
