```python
# Cuántos números son pares en el arreglo?
numeros = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5, 8, 9, 7, 9, 3, 2]

n_pares = 0

for numero in numeros:
    if numero % 2 == 0:
        n_pares += 1

print(f"Número de elementos pares: {n_pares}")
```

```python
# Asuman que el arreglo tiene al menos un elemento
# Hallar el mayor valor en el arreglo números
numeros = [31, 41, 59, 26, 53, 58, 97, 93, 2]
maximo = numeros[0]
for numero in numeros:
    if numero > maximo:
        maximo = numero
print(f"Máximo valor: {maximo}")
```

```python
print(f"Máximo valor: {max(numeros)}")
```
