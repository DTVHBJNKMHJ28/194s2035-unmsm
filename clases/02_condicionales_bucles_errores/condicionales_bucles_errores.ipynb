```python
import random

suerte = 2

lanzamiento = -1

mayor_lanzamiento = -1
numero_de_lanzamientos = 0

while lanzamiento != suerte:
    numero_de_lanzamientos += 1
    lanzamiento = random.randint(1, 7)
    print(f"Intento #{numero_de_lanzamientos}: {lanzamiento}")

    if mayor_lanzamiento < lanzamiento:
        mayor_lanzamiento = lanzamiento

print(f"Numero de la suerte: {lanzamiento}")
print(f"Mayor lanzamiento obtenido: {mayor_lanzamiento}")
print(f"NÃºmero de intentos: {numero_de_lanzamientos}")
```

```python
# numero_digitos(n) que retorne el nÃºmero de dÃ­gitos de un entero positivo
# numero_digitos(5) -> 1
# numero_digitos(123) -> 3
# numero_digitos(123123123) -> 9
```

```python
def numero_digitos(n):
    if n < 10:
        return 1
    return 1 + numero_digitos(n // 10)
```

```python
print(numero_digitos(5))
print(numero_digitos(123))
print(numero_digitos(123123123))
```

```python
def numero_digitos_gabriel(n):
    if n < 10:
        return 1
    if n < 100:
        return 2
    if n < 1000:
        return 3
```

```python
print(numero_digitos_gabriel(5))
print(numero_digitos_gabriel(123))
print(numero_digitos_gabriel(123123123))
```

```python
def numero_digitos_laurente(n):
    digitos = 0
    while n >= 10:
        digitos += 1
        n = n // 10
    return 1 + digitos
```

```python
print(numero_digitos_laurente(5))
print(numero_digitos_laurente(123))
print(numero_digitos_laurente(123123123))
```

```python
def dividir(numerador, denominador):
    if denominador == 0:
        raise Exception("El denominador es cero :(")
    return numerador / denominador

print(dividir(10, 2))
print(dividir(50, 20))
try:
    print(dividir(1, 0))
except Exception as e:
    print(e)
```

```python
alumnos = ["alessandro", "alexander", "bryam", "daira"]

def obtener_alumno(indice):
    if indice < 0:
        raise Exception("El Ã­ndice tiene que ser no negativo")
    return alumnos[indice]

print(obtener_alumno(2))
print(obtener_alumno(3))
try:
    print(obtener_alumno(-1))
except IndexError:
    print("El Ã­ndice excede la cantidad de alumnos")
except Exception as e:
    print(e)
```
